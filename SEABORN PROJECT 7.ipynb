```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.set(style = "ticks", color_codes = True)
```

```python
tips = sns.load_dataset("tips")
sns.catplot( x = "day", y = "total_bill", data = tips)
```

```python
# jitter parameter 

sns.catplot( x= "day", y = "total_bill", jitter = False,data = tips)
```

```python

sns.catplot( x= "day", y = "total_bill", kind = "swarm", data = tips)
```

```python
# hue


sns.catplot( x= "day", y = "total_bill",hue = "sex", kind = "swarm", data = tips)
```

```python

sns.catplot( x= "size", y = "total_bill",hue = "sex", kind = "swarm", data = tips)
```

```python
# create order


sns.catplot( x= "smoker", y = "tip",order = ["No", "Yes"], data = tips)
```

```python
# swap variables



sns.catplot( x= "total_bill", y = "day",kind = "swarm", data = tips)
```

```python
# dist of observations within categories

# boxplots
# whiskers and box
# IQR
# outliers



sns.catplot( x= "day", y = "total_bill",kind = "box", data = tips)
```

```python
# dodging

tips["weekend"] = tips["day"].isin(["Sat", "Sun"])

sns.catplot(x = "day", y = "total_bill", hue = "weekend",
            kind = "box", dodge = False, data = tips)
```

```python
# countplot
sns.catplot( x = "deck", kind = "count", palette = "ch:25", data = titanic)
```

```python
sns.catplot( x = "deck", hue = "class",  kind = "count", palette = "pastel",edgecolor = ".6", data = titanic)
```

```python
# pointplot
sns.catplot( x = "sex", y = "survived",  hue = "class", kind = "point",  data = titanic)
```
